```python
import json
import numpy as np
import matplotlib.pylab as plt
from matplotlib.animation import FuncAnimation
import os
from visual import *
```

```python
def create_config(file_name, t_left, t_right, N, solver_type, DEBUG=0):
    data = {
        't_left': float(t_left),
        't_right': float(t_right),
        'N': int(N),
        'solver_type': solver_type,
        'DEBUG': DEBUG,
    }

    with open(''.join(['configs/', file_name]), 'w') as outp:
        json.dump(data, outp)
```

```python
def read_config(file_name):
    data = dict()
    with open(''.join(['configs/', file_name]), 'r') as inp:
        data = json.load(inp)
    
    return data
```

```python
create_config('config200H.json', 0, 5, 200, 'H')
```

```python
def run_sim(config_files):
    outp_files = []
    for file in config_files:
        data = read_config(file)
        vals = list(map(str, data.values()))

        outp_name = "data"+str(data['N'])+data['solver_type']+'.txt'
        outp_files.append(outp_name)
        os.system("./harmonic_solver "+" ".join(vals)+" >> "+"data/"+outp_name)

    return outp_files
```

```python
configs = ['config50H.json', 'config100H.json', 'config150H.json', 'config200H.json']
datas = []
```

```python
datas = run_sim(configs)
```
