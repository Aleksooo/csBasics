```python
def create_config(**kwargs):
    file_id = ''
    for name, val in kwargs.items():
        file_id += '_'"{:1.1f}".format(val)
    with open('configs/'+'config'+file_id+'.json', 'w') as outp:
        json.dump(kwargs, outp)

    return file_id
```

```python
# Creating config for pendulum

ids = [create_config(left = 0, right = 2, N = 5000, x0 = 0, v0 = 1, OMEGA = 1, GAMMA = 0.1, A = 0.5, BIG_OMEGA = 2.5)]
```
